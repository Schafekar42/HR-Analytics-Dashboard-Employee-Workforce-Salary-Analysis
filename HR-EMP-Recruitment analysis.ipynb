# HR Analytics : Workforce Insights & Employee Trends

 ## Business Problem

The HR department lacks a centralized system to monitor key workforce metrics such as employee headcount, gender diversity, salary distribution, department-wise employee strength, and employee attrition. As a result, decision-making related to workforce planning, budgeting, and employee management becomes time-consuming and less effective.

## Project Objective

The objective of this project is to develop an interactive HR Analytics Dashboard that provides actionable insights into workforce data. The dashboard helps HR managers and business leaders:

 * Monitor total headcount, active employees, and terminated employees.

* Analyze employee distribution across departments and locations.

* Track gender diversity within the organization.

* Evaluate salary distribution by department and employee groups.

* Identify workforce trends to support strategic HR decision-making.

* Improve reporting efficiency through interactive visualizations and real-time insights.

```python
import ipywidgets as widgets
from IPython.display import display

def plot_department(year):

    filtered_df = HR_Df[HR_Df['year'].dt.year == year]

    Dept_Count = filtered_df.groupby('Department')['EmployeeID'].count()

    ax = Dept_Count.plot(kind='bar')
    ax.bar_label(ax.containers[0])
    plt.title(f'Employee Count by Department - {year}')
    plt.xlabel('Department')
    plt.ylabel('Employees')
    plt.xticks(rotation = 0)
    plt.show()

year_dropdown = widgets.Dropdown(
    options=sorted(HR_Df['year'].dt.year.unique()),
    description='Year'
    
)

widgets.interactive(plot_department, year=year_dropdown)
```

```python
import ipywidgets as widgets
from IPython.display import display

def plot_PerformanceScore(year):

    filtered_df = HR_Df[HR_Df['year'].dt.year == year]

    Performance_Count = filtered_df['PerformanceScore'].value_counts()

    plt.figure(figsize=(8,6))
    plt.pie(
        Performance_Count,
        labels=Performance_Count.index,
        autopct='%1.1f%%'
    )
    plt.title(f'Year Wise Employee Performance - {year}')
    plt.show()

year_dropdown = widgets.Dropdown(
    options=sorted(HR_Df['year'].dt.year.dropna().unique()),
    description='Year:'
)

interactive_plot = widgets.interactive(
    plot_PerformanceScore,
    year=year_dropdown
)

display(interactive_plot)
    
```
